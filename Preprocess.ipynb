```python
import pandas as pd
import nltk
```

```python
df1 = pd.read_csv('star_1.csv',encoding='utf-8')
df2 = pd.read_csv('star_2.csv',encoding='utf-8')
df3 = pd.read_csv('star_3.csv',encoding='utf-8')
df4 = pd.read_csv('star_4.csv',encoding='utf-8')
frame = [df1,df2,df3,df4]
df = pd.concat(frame)
df.head()
```

```python
df = df.reset_index(drop=True)
df = df.dropna()
df
```

```python
#去除文本中乱码
for i, row in df.iterrows():
    comment = str(row['Comment'])
    cleaned_comment = comment.encode('ascii', 'ignore').decode('ascii')
    df.at[i, 'Comment'] = cleaned_comment
```

```python
#文本转小写
df['Comment'] = df['Comment'].str.lower()
```

```python
#去除所有标点符号
import re
from string import punctuation as punctuation_en

punctuation = punctuation_en

def remove_punctuation(text):
    return text.translate(str.maketrans('','',punctuation))

df['Comment'] = df['Comment'].apply(remove_punctuation)
```

```python
import nltk.tokenize as tk

tokenizer = tk.WordPunctTokenizer()
data = []
for i,context in df.iterrows():
    comment = context['Comment']
    comment = tokenizer.tokenize(comment)
    data.append(comment)
    print(comment)
```

```python
word_num = []
for i in data:
    count = len(i)
    word_num.append(count)
word_num
```

```python
# NLTK Stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

stop_words.extend(['from', 'subject', 're', 'edu', 'use', 'not', 'would', 'say', 
                   'could', '_', 'be', 'know', 'good', 'go', 'get', 'do', 'done', 'try', 
                   'many', 'some', 'nice', 'thank', 'think', 'see', 'rather', 'easy', 'easily', 
                   'lot', 'lack', 'make', 'want', 'seem', 'run', 'need', 'even', 'right', 'line', 
                   'even', 'also', 'may', 'take', 'come'])

content_words_list = []
for context in data:
    words = []
    for word in context:
        if word not in stop_words:
            words.append(word)
    content_words_list.append(words)
content_words_list
```

```python
#将词数与词表合并并保存到一张csv里
num = pd.DataFrame(word_num)

corpus = []
for content in content_words_list:
    str_content = " ".join(content)
    corpus.append(str_content)
corpus = pd.DataFrame(corpus)

stars = df['Star']
date = df['Date']
helpfulness = df['Helpfulness']

frame = [date,num,corpus,stars,helpfulness]
comment_low_stars = pd.concat(frame,axis=1)
comment_low_stars.columns = ['date','length','corpus','stars','helpfulness']
```

```python
comment_low_stars.to_csv('comment_5_stars_2.csv')
```
